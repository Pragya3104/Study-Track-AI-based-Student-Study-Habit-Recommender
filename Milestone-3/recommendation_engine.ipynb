# Milestone 3: Recommendation Engine

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../Milestone-2/customer_clusters.csv")

```

```python
# Use the cluster result from milestone 2

print(df.head())
```

```python
# Recommendation function
def recommend(cluster):
    if cluster == 1:
        return "Provide loyalty rewards & exclusive offers"
    elif cluster == 0:
        return "Send discount coupons to increase engagement"
    elif cluster == 2:
        return "Recommend premium or curated product bundles"
    else:
        return "General engagement follow-up"

df["Recommendation"] = df["Cluster"].apply(recommend)

df["Suggested_Tool"] = df["Cluster"].map({
    1: "Loyalty Membership / Cashback Program",
    0: "Email Discount Campaigns / Festival Offers",
    2: "Personalized Product Recommendations / Premium Catalog"
})

df.head()
```

```python
# Visualization: Recommendation Count Plot
sns.countplot(x="Recommendation", data=df, palette="Set2")
plt.xticks(rotation=20)
plt.title("Recommendation Distribution per Cluster")
plt.savefig("visualizations/recommendation_countplot.png")
plt.show()
```

```python
# Save Final Dataset
df.to_csv("customer_recommendations.csv", index=False)
print("File saved as customer_recommendations.csv")
```
